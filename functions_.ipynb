6. Explain the concept of generators in Python and how they are defined.
 -In python,generators are a type of iterable,like lists or tuples,that allow you to produce a sequence of values lazily (one at a time) instead of computing them all at once and storing them in memory.this makes them more memory-efficient,especially when working with large datasets or infinite sequences.
 - a generator function is a regular function that uses the yield statement to produce values one at a time.the yield statement pauses the function,saving its state.when next()is called on the generator,execution resumes from where it left off.the function stops when there are no more yield statements or when it explicity ends with return.


```python
def square_number(n):
  result =[]
  for i in n:
    result.append(i**2)
  return result
square_number([5,10,15])
```

```python
#yield
def square_number(n):
  for i in range(n):
    yield i**2
square_number(10)
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.


```python
def celsius_to_fahrenheit(celsius):
  return (celsius * 9/5) + 32
celsius_temperature = [0,20,30,40,100]
fahrenheit_temperature = list(map(celsius_to_fahrenheit,celsius_temperature))
print("celsius temperature:",celsius_temperature)
print("fahrenheit temperaure:",fahrenheit_temperature)
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def remove_vowels(input_string):
  def is_not_vowels(char):
    vowels = "aeiouAEIOU"
    return char not in vowels
  result = filter(is_not_vowels,input_string)
  return "".join(result)
input_string = input("enter a string:")
output_string = remove_vowels(input_string)
print("string without vowels:",output_string)
```
